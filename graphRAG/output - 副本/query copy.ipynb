```python
# 修改 Cell 10 代码，添加过滤逻辑

from pyvis.network import Network

net = Network(height="1200px", width="100%", notebook=True, cdn_resources="in_line")

# 初始化颜色映射字典
level_color_map = {"核心": "#FF6B6B", "普通": "#4ECDC4"}  # 固定层级色系
community_palette = ['#0000FF',  '#0033FF',  '#0066FF','#0099FF',  '#00CCFF', '#00FFFF', '#33B5E5' '#FF0000',  '#FF3333', '#FF6666', '#FF9999', '#FFCCCC', '#FFE5E5',  '#FFB3B3',  '#FF8080', '#FF4D4D', '#FF1A1A','#CC0000']  # 社区色板（可扩展）

# 定义需要筛选的生态子系统关键词
ecological_subsystems = ["森林生态", "湿地生态", "草原生态", "荒漠生态", "农田生态", "城市生态", "海洋生态"]

# 存储相关节点的集合
relevant_nodes = set()

# 第一次遍历：找到所有包含生态子系统关键词的节点
for _, row in entities.iterrows():
    node_title = row["title_x"]
    # 检查节点名称是否包含任何生态子系统关键词
    is_relevant = any(subsystem in node_title for subsystem in ecological_subsystems)
    if is_relevant:
        relevant_nodes.add(node_title)

# 第二次遍历：添加与生态子系统直接相连的节点
for _, rel in relationships.iterrows():
    source = rel['source']
    target = rel['target']
    if source in relevant_nodes:
        relevant_nodes.add(target)
    if target in relevant_nodes:
        relevant_nodes.add(source)

# 遍历实体数据配置颜色，仅添加相关节点
for _, row in entities.iterrows():
    node_title = row["title_x"]
    
    # 跳过不相关的节点
    if node_title not in relevant_nodes:
        continue
        
    # 判断层级优先级
    if row["level_x"] in level_color_map:
        color = level_color_map[row["level_x"]]
    else:
        # 按社区ID循环取色（避免超出色板范围）
        comm_idx = row["community"] % len(community_palette)
        color = community_palette[comm_idx]
    
    # 添加节点
    net.add_node(
        n_id=node_title,
        label=node_title,
        title=f"层级：{row['level_x']}\n社区：{row['summary']}",
        color=color,
        group=row["community"]  # 可选：用于分组交互
    )
```

```python
# 修改 Cell 12 代码，只添加相关节点之间的边
for _, rel in relationships.iterrows():
    source = rel['source']
    target = rel['target']
    
    # 仅当源节点和目标节点都在相关节点集合中时才添加边
    if source in relevant_nodes and target in relevant_nodes:
        net.add_edge(
            source=source,
            to=target,
            value=rel['weight'],  # 边粗细映射权重
            title=f"关系类型：{rel['description']}\n权重：{rel['weight']:.2f}",
            color  = "#888"          # 原有边灰色
        )
```

```python
# 修改 Cell 14 代码，只添加相关节点之间的新边
for _, row in new_edges_df.iterrows():
    src = str(row["source"])
    tgt = str(row["target"])
    
    # 仅当源节点和目标节点都在相关节点集合中时才添加新边
    if src in relevant_nodes and tgt in relevant_nodes:
        rel = row["relation"]
        sim = float(row["similarity"])
        net.add_edge(
            source = src,
            to     = tgt,
            value  = 1,  # 可视化宽度
            title  = f"新补全：{rel}\n相似度：{sim:.2f}",
            color  = "#FF9900",
            width  = 3,
            smooth = False
        )
```

## 度量

```python
import pandas as pd
import networkx as nx

# 读取 Parquet 文件（假设在当前工作目录）
nodes_df = pd.read_parquet("create_final_nodes.parquet")
relationships_df = pd.read_parquet("create_final_relationships.parquet")
community_df = pd.read_parquet("create_final_community_reports.parquet")
```

```python

# 构建有向图
G = nx.DiGraph()
# 添加节点并附加社区信息
for idx, row in nodes_df.iterrows():
    entity = row['title']  # 假设节点表中实体字段名为 'entity'
    # 从 community_df 获取该实体的社区编号和摘要
    community_info = community_df[community_df['title'] == entity]
    if not community_info.empty:
        comm_id = int(community_info['community_id'].values[0])
        summary = community_info['summary'].values[0]
    else:
        comm_id = -1
        summary = ""
    G.add_node(entity, community=comm_id, summary=summary)

# 添加边及其属性
for idx, row in relationships_df.iterrows():
    src = row['source']      # 假设关系表中源实体字段名为 'source'
    tgt = row['target']      # 目标实体字段名为 'target'
    desc = row['description']  # 关系描述字段
    weight = row.get('weight', None)  # 如果存在 weight 字段
    G.add_edge(src, tgt, description=desc, weight=weight)
```

```python
import matplotlib.pyplot as plt

# 节点入度和出度
in_degrees = dict(G.in_degree())
out_degrees = dict(G.out_degree())

# 转换为列表便于绘图
in_vals = list(in_degrees.values())
out_vals = list(out_degrees.values())
```

```python
# 绘制入度分布直方图
plt.rc('font', family='SimHei')
plt.figure(figsize=(6,4))
plt.hist(in_vals, bins=30, color='skyblue', edgecolor='black')
plt.title("节点入度分布")
plt.xlabel("入度")
plt.ylabel("节点数量")
plt.show()
```

```python
# 绘制出度分布直方图
plt.figure(figsize=(6,4))
plt.hist(out_vals, bins=30, color='lightgreen', edgecolor='black')
plt.title("节点出度分布")
plt.xlabel("出度")
plt.ylabel("节点数量")
plt.show()
```

```python
# 聚类系数（使用无向图近似）
undirected_G = G.to_undirected()
clustering_dict = nx.clustering(undirected_G)
clust_vals = list(clustering_dict.values())
avg_clustering = nx.average_clustering(undirected_G)
print(f"平均聚类系数: {avg_clustering:.4f}")
```

```python
# 绘制聚类系数直方图
plt.figure(figsize=(6,4))
plt.hist(clust_vals, bins=30, color='salmon', edgecolor='black')
plt.title("节点聚类系数分布")
plt.xlabel("聚类系数")
plt.ylabel("节点数量")
plt.show()
```

```python
# 弱连通分量
weakly_components = list(nx.weakly_connected_components(G))
num_components = len(weakly_components)
largest_comp_size = max(len(c) for c in weakly_components)
print(f"弱连通分量数量: {num_components}")
print(f"最大连通分量大小: {largest_comp_size}")
```

```python
# 图密度
density = nx.density(G)
print(f"图的密度 (edge density): {density:.6f}")
```

---

### 1. 稀疏性与孤立性诊断

* 从入/出度直方图可见 >70 % 节点 (deg = 0 或 1)，且 **入度 = 0** 节点数量最多 → 典型“尾实体孤立”现象。
* 聚类系数大批节点为 0，说明它们既不是三角闭环的一员，也缺少邻居互连；这些节点极易在推理时被模型忽略，需要补全。
* 18 个弱连通分量中，最大分量已囊括主干知识，但仍有 17 个“散片”子图：往往是条款型概念与正文指标未被 GraphRAG 正确对齐。

> **可操作策略**
>
> 1. **实体对齐**：利用名称或同义词检索将孤立子图合并到主分量；
> 2. **横向关系生成**：以低度节点为头实体，让 LLM 生成 “影响/从属/计算自” 等关系 → 连接回入度高的核心指标。

### 2. Prompt-Tuning Few-Shot 设计建议

| 步骤            | why                            | how                                                |
| ------------- | ------------------------------ | -------------------------------------------------- |
| **Anchor 选取** | 高入度节点是知识汇聚点，提供上下文最丰富           | 在 Few-Shot 示例中固定高入度节点 (`deg_in ≥ 4`) 作为头实体，询问缺失尾实体 |
| **关系优先级**     | 当前图谱边大多为“定义—指标”，缺少“影响”“组成”“度量” | 手动或自动统计关系频率后，针对低频关系设计 Soft Prompt，并插入 1-3 条示例      |
| **孤立节点重连**    | 出/入度为 0 节点容易被忽略                | 针对孤立节点构造问句：`"<孤立指标> 影响了哪些生态服务？"` 交由 LLM 生成尾实体      |
| **评估集**       | 需体现连通性提升                       | 预留部分“跨社区”真实三元组做 blind-test；Hits\@1/10 改进越大说明连通性越好  |

### 3. 后续补全效果衡量

> **目标**：连通分量数 ↓、平均度 ↑、密度 ↑、平均最短路径 ↓
>
> 1. 复跑上述指标，观察 **最大连通分量占比是否提升**（理想 >85 %）。
> 2. 当平均聚类系数显著提升而密度仍保持稀疏，说明新边更集中于局部功能簇 —— 满足“意义性”而非随机连边。
> 3. 若密度 >0.02 且聚类系数维持 \~0.25-0.35，则表示既提升了横向连通，又避免过度冗余。

## 图结构状态一句话总结

> 该生态标准知识图谱呈现“**稀疏星簇+孤立小片**”结构：少量核心定义是入度中心，大批指标叶节点横向断裂，导致全图密度仅 0.7 %。通过 Prompt-Tuning Few-Shot 补全，优先为度=0/1 节点生成跨章节“影响/组成”类关系，可在不引入噪声的前提下把 18 个弱连通分量收敛到单一主分量，并让平均聚类系数与 Hits\@K 指标同步提升。


## 补全后度量

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# —— 1. 读取数据 —— 
nodes_df = pd.read_parquet("create_final_nodes.parquet")
relationships_df = pd.read_parquet("create_final_relationships.parquet")
community_df = pd.read_parquet("create_final_community_reports.parquet")
new_edges_df = pd.read_csv("new_edges.csv", encoding="utf-8-sig")
```

```python
# —— 2. 构建原始图 & 补全后图 —— 
G_orig = nx.DiGraph()
for _, row in nodes_df.iterrows():
    G_orig.add_node(str(row['title']))
for _, row in relationships_df.iterrows():
    G_orig.add_edge(str(row['source']), str(row['target']))

G_new = G_orig.copy()
for _, row in new_edges_df.iterrows():
    G_new.add_edge(str(row['source']), str(row['target']))
```

```python
# —— 3. 计算指标函数 —— 
def compute_metrics(G):
    in_deg  = list(dict(G.in_degree()).values())
    out_deg = list(dict(G.out_degree()).values())
    und     = G.to_undirected()
    clust   = list(nx.clustering(und).values())
    return {
        "in_degree"  : in_deg,
        "out_degree" : out_deg,
        "clustering" : clust,
        "avg_cluster": nx.average_clustering(und),
        "num_comps"  : len(list(nx.weakly_connected_components(G))),
        "max_comp"   : max(len(c) for c in nx.weakly_connected_components(G)),
        "density"    : nx.density(G)
    }
```

```python
m0 = compute_metrics(G_orig)
m1 = compute_metrics(G_new)

# —— 4. 打印对比结果 —— 
print(f"{'指标':<20}{'原始':>10}{'补全后':>15}")
print(f"{'平均聚类系数':<20}{m0['avg_cluster']:>10.4f}{m1['avg_cluster']:>15.4f}")
print(f"{'弱连通分量数':<20}{m0['num_comps']:>10}{m1['num_comps']:>15}")
print(f"{'最大连通分量':<20}{m0['max_comp']:>10}{m1['max_comp']:>15}")
print(f"{'图密度':<20}{m0['density']:>10.6f}{m1['density']:>15.6f}")
```

```python
import random
# 按关系类型分组抽样
relation_to_triples = {}
for h, r, t in triples:
    relation_to_triples.setdefault(r, []).append((h, r, t))

few_shot_samples = []
for r, trs in relation_to_triples.items():
    samples = random.sample(trs, min(5, len(trs)))  # 每种关系最多5条
    few_shot_samples.extend(samples)
```

```python
# 构造 Prompt 数据集
prompts = []
labels = []
for h, r, t in few_shot_samples:
    # (头, 关系, ?) -> 填补尾实体
    prompt = f"{h}{r}什么？"  # 简单示例：如 "张三喜欢什么？"
    label = t
    prompts.append(prompt)
    labels.append(label)
    # (?, 关系, 尾) -> 填补头实体
    prompt2 = f"什么{r}{t}？"  # 如 "什么喜欢苹果？"
    label2 = h
    prompts.append(prompt2)
    labels.append(label2)
```

```python
from sklearn.model_selection import train_test_split
# 划分训练和测试集
train_texts, test_texts, train_labels, test_labels = train_test_split(prompts, labels, test_size=0.2, random_state=42)
```
